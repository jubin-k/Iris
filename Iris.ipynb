```python
#removing the column 'Id' because it is independant on the output
df=df.drop(['Id'],axis=1)
df
```

```python
#checking the count of each outputs
df1=df.groupby('Species')['Species'].count()
df1
```

```python
#seperating input under a variable 'x'
x=df.iloc[:,:-1].values
x
```

```python
#seperating output under a variable 'y'
y=df.iloc[:,-1].values
y
```

```python
#splitting the data into train and test data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30)
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
#normalizing the values to minimize errors 
#normalization is done only to input datas because it is a classification problem
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
x_test
```

```python
#creating the ML model and testing the test data to the train data
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=11)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
y_pred
```

```python
#predicting the flower category using the given inputs
print(knn.predict([[4.6,3.2,3.8,1.8]]))
```

```python
#we need to plot Confusionmatrix to compare y_test with y_pred
#define categories in a variable
from sklearn.metrics import confusion_matrix,accuracy_score,ConfusionMatrixDisplay
label=['Iris-setosa','Iris-versicolor','Iris-virginica']
mat=confusion_matrix(y_pred,y_test)
cmd=ConfusionMatrixDisplay(mat,display_labels=label)
cmd.plot()
mat
```

```python
#calculating the accuracy score
score=accuracy_score(y_pred,y_test)
score
```
